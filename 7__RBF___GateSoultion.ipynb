```python
import numpy as np
import sys
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.svm import SVC
import warnings
from math import e
plt.style.use("seaborn-talk")
%matplotlib notebook
warnings.filterwarnings("ignore")
```

```python
INPUT = np.array([[0, 0], [1, 0], [0, 1], [1, 1]])
XOR = np.array([0,1,1,0])
OR = np.array([0,1,1,1])
AND = np.array([0,0,0,1])
epoch = 100
k = "rbf"
xor_model = SVC(kernel=k,C=epoch).fit(INPUT,XOR)
or_model = SVC(kernel=k,C=epoch).fit(INPUT,OR)
and_model = SVC(kernel=k,C=epoch).fit(INPUT,AND)


X = [0,0,1,1]
Y = [0,1,0,1]
fig = plt.figure(figsize=(13,6))
ax = plt.subplot(121)
ax.scatter(X,Y,s=300,c=[0,0,0,1])
ax.plot([0.1,1],[1,0.1])
ax.set_title("AND")
ax = plt.subplot(122)
ax.scatter(X,Y,s=300,c=[0,1,1,0])
ax.set_title("XOR")
```

```python
def and_gate(x, y):
    return and_model.predict([[x,y]])[0]


def or_gate(x, y):
    return or_model.predict([[x,y]])[0]


def xor_gate(x, y):
    return xor_model.predict([[x,y]])[0]
```

```python
%matplotlib notebook
INPUT = np.array([[1, 0, 0], [1, 1, 0], 
                  [1, 0, 1], [1, 1, 1]])

fig = plt.figure(figsize=(10, 8))

Input = [[0, 0], [0, 1], [1, 0], [1, 1]]

ax = plt.subplot(221, projection="3d")

ax.scatter(INPUT[:, 1:2], INPUT[:, 2:3], AND, 
           s=20, c='g', label="Data points")
ax.set(
    facecolor="white",
    title="AND Gate"
)
X = np.arange(-2, 2, 0.5)
X, Y = np.meshgrid(X, X)
Z = np.array([[and_gate(z, z) for z in i] for i in X])
ax.plot_surface(X, Y, Z, alpha=0.2, color="r")


ax = plt.subplot(222, projection="3d")
ax.scatter(INPUT[:, 1:2], INPUT[:, 2:3], OR, 
           s=20, c='g', label="Data points")
ax.set(
    facecolor="white",
    title="OR Gate",
)
X = np.arange(-2, 2, 0.5)
X, Y = np.meshgrid(X, X)
Z = np.array([[or_gate(z, z) for z in i] for i in X])
ax.plot_surface(X, Y, Z, alpha=0.2, color="r")


ax = plt.subplot(223, projection="3d")
ax.scatter(INPUT[:, 1:2], INPUT[:, 2:3], [0, 1, 1, 0], 
           s=20, c='g')
ax.set(
#     facecolor="white",
    title="XOR Gate"
)
X = np.arange(-2, 2, 0.5)
X, Y = np.meshgrid(X, X)
Z = np.array([[xor_gate(z, z) for z in i] for i in X])
ax.plot_surface(X, Y, Z, alpha=0.2, color="r")
plt.subplots_adjust(top=1, bottom=0, left=-0.01)
```
